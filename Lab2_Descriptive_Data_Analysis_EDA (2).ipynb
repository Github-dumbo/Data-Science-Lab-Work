```python
import pandas as pd


df_txt = pd.read_csv(
    "Facebook Friends.txt",
    delim_whitespace=True,
    encoding="cp1252"
)

df_csv = pd.read_csv("Facebook Friends.csv")
df_xlsx = pd.read_excel("Facebook Friends.xlsx")

print("TXT shape:", df_txt.shape)
print("XLSX shape:", df_xlsx.shape)
print("CSV shape:", df_csv.shape)
```

```python
#The three file types produced different dataframe structures.
#The excel file preserved the correct formatting.
#The csv file lost delimiter structure, resulting in 3 columns.
#The txt file required whitespace parsing but produced inconsistent structure.
#Hence, file formatting and encoding singnificantly affect how data is interpreted.
```

```python
df = df_xlsx.copy()
```

```python

print(df.head())
print(df.tail())
print(df.columns)
print(df.dtypes)
print(df.info())
print(df.isnull().sum())
```

```python
df.describe()
```

```python
#The describe() function generates descriptive statistics for all numerical variables in the dataset.
#It summarizes the distribution, central tendency, and variability of each variable.
```

```python
#Each column represents a variable (e.g., Age, Photos, Friends).
#Each row represents a statistical measure.

#For example, in the Friends column:

#Mean (~695) → Average number of friends

#Median (~563) → Middle value when sorted

#Max (4999) → Highest friend count

#Std (~600) → Indicates high variability

#If the mean is much larger than the median, the data is likely right-skewed.
```

```python
#Key Statistical Attributes Indicated
#Central Tendency
#Mean shows the overall average.
#Median (50%) shows the middle value.
#Comparing mean vs median helps detect skewness.
```

```python
#Variability (Spread)
#Standard deviation (std) shows how spread out the values are.
#A large std (like Photos or Friends) indicates high variation between users.
```

```python
#Range
#Min and Max show the range of the data.
#Large differences between min and max suggest outliers.
```

```python
#Quartiles
#25%, 50%, 75% divide the data into four equal parts.
#Helps understand distribution without being affected by outliers.
```

```python
import matplotlib.pyplot as plt

df.hist(figsize=(12,10))
plt.show()
```

```python
df["Friends"].hist(bins=20)
plt.title("Distribution of Number of Friends")
plt.xlabel("Number of Friends")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Gender "].value_counts().plot(kind="bar")
plt.title("Gender Frequency")
plt.show()
```

```python
#The Gender frequency plot helps visualize the distribution of categorical data.
#It shows how observations are distributed across gender categories and reveals whether the dataset is balanced or dominated by a particular group.
```

```python
#Data visualization transforms numerical data into graphical form, making patterns, trends, and irregularities easier to detect than through summary statistics alone.

#While descriptive statistics provide averages and spread, visualizations reveal:

#The shape of distributions

#The presence of skewness

#Potential outliers

#Patterns of concentration

#Differences between categories

#For example, a mean value does not show whether the data is normally distributed or heavily skewed, but a histogram does.
```

```python
#Insights that can be gained from vizualisation -
#Distribution Shape
#Variability in Activity
#Detection of Outliers
#Categorical Balance
```

```python
#Summary
#This analysis of the Facebook data was performed using descriptive statistics and visualisation.
#Results show that the activity variables Photos, Tags, and Friends are strongly right skewed, indicating a small number of vastly more active users.
#The Age variable indicates that users are mostly young.
#Visualisations showed spread and outliers not seen in summary stats, showing engagement and connectivity patterns.
```
