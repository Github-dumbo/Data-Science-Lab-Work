```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 3 * np.pi ,0.1) #(starting, ending, interval)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel("x axis label")
plt.ylabel("Trigonometric Function")
plt.title("Sine And Cosine Function")
plt.show()
```

```python
# subplot

plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title("Sine Function")

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title("Cosine Function")
plt.xlabel("Variable x")
plt.ylabel("Cosine")
plt.show()
```

```python
p = np.arange(0,10,1)
q =p**2
d = np.exp(p)

plt.scatter(p,q)
plt.scatter(p,d/1000)
```

```python
m = np.arange(0, 10, 1)
l_m = 3*m + 2
l_e = np.exp(m)
l_sin = np.sin(m)
l_cos = np.cos(m)


plt.figure()
plt.suptitle("Figure 1")

plt.subplot(2,2,1)
plt.plot(m, l_m)
plt.title("Linear Function (3x + 2)")

plt.subplot(2,2,2)
plt.plot(m, l_e)
plt.title("Exponential Function")

plt.subplot(2,2,3)
plt.plot(m, l_sin)
plt.title("Sin Function")

plt.subplot(2,2,4)
plt.plot(m, l_cos)
plt.title("Cos Function")
plt.show()
```
